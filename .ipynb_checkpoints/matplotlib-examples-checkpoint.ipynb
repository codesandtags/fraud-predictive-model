## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Simple Plot",
      xlabel = "X axis",
      ylabel = "Y Axis")

# 5. Save & show (you save the whole figure)
fig.savefig('./images/sample-plot.png')
```

## Making figures with Numpy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another example with scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from a dictionary
nut_butter_prices = { "Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title= 'Edwins butter prices',
      xlabel='Butter type',
      ylabel='Price ($)');
```

```python
# Make a plot from a dictionary but in an horizontal way
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title= 'Edwins butter prices',
      xlabel='Butter type',
      ylabel='Price ($)');
```

```python
# Make a histogram and plot it
x = np.random.randn(100)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols=2, figsize = (10, 5))

# Plot each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.randn(10), np.random.rand(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(100));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows = 2, ncols=2, figsize = (10, 5))

# Plot each different axis
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.randn(10), np.random.rand(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(100));
```
